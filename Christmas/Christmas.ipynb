```python
minor = christmas[christmas['mode'].str.contains("minor", na=False)]
```

```python
minor = minor.drop(columns=['Unnamed: 0', 'track_uri', 'duration_ms', 'time_signature', 'playlist_name', 'playlist_img', 'album_img', 'track_name', 'artist_name', 'album_name'])
```

```python
minor.head()
```

```python
minorheat = minor[['danceability', 'tempo', 'energy', 'acousticness']]
sns.heatmap(minorheat.corr(),annot=False,cmap="cool")
```

```python
major = christmas[christmas['mode'].str.contains("major", na=False)]
```
